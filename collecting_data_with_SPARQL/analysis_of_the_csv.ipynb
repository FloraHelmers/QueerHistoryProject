## in a graph form 

```python
Graph_gay_graph = nx.from_pandas_edgelist(gay_graph, source="subject", target="object", edge_attr="predicate")
```

```python
nx.draw(Graph_gay_graph)
#a ne pas relancer
```

```python
connect_comp = nx.connected_components(Graph_gay_graph)
for c in connect_comp:
    print(len(c))
```
